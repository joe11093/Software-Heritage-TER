```python
#MODULE IMPORTS
import json
from time import time
from datetime import datetime
import launchpadlib
from launchpadlib.launchpad import Launchpad

#CONSTANTS
cachedir = "/home/anonbnr/.launchpadlib/cache/" #cache directory
project_attributes_path = 'project_attributes.txt'

#logging in anonymously to the API (read-only)
launchpad = Launchpad.login_anonymously('just testing', 'production', cachedir, version='devel')
```

```python
#retrieving the list of launchpad projects and their metadata
projects = launchpad.projects
```

```python
#retrieving a sample of launchpad projects containing
#projects using git, bzr or no vcs
git_counter = 50
bzr_counter = 50
sample = []
print("Started retrieving projects at {}".format(datetime.now()))
start_time = time()
for project in projects:
    if project.vcs == 'Git' and git_counter != 0:
        sample.append(project)
        git_counter -= 1
    elif project.vcs == 'Bazaar' and bzr_counter != 0:
        sample.append(project)
        bzr_counter -= 1
print("Finished retrieving projects at {} (took {} seconds)".format(datetime.now(), time() - start_time))
```

```python
#Displaying attribute values for git and bazaar projects
for project in sample:
    print("""
    Date created: {}
    Description: {}
    Display Name: {}
    Download URL: {}
    Homepage URL: {}
    Name: {}
    self_link: {}
    Title: {}
    VCS: {}
    \n
    """.format(project.date_created, project.description, project.display_name, 
               project.download_url, project.homepage_url, 
               project.name, project.self_link, project.title, project.vcs))
```

```python
#retrieving the list of projects using git and their metadata
git_projects = []

print("Started retrieving git projects at {}".format(datetime.now()))
start_time = time()
#getting git projects
for project in projects:
    if project.vcs == 'Git':
        git_projects.append(project)
print("Finished retrieving projects at {} (took {} seconds)".format(datetime.now(), time() - start_time))
```

```python
def get_git_repos():
    for project in git_projects:
        yield launchpad.git_repositories.getRepositories(target=project.self_link)

def export_git_repos_json(filename):
    with open(filename, 'a') as file:
        list_json = []
        for git_repos in get_git_repos():
            for git_repo in git_repos.entries:
                if git_repo != None:
                    list_json.append(git_repo)
        json.dump(list_json, file, indent=4)

def import_git_repos_json(filename):
    with(open(filename, 'r')) as file:
        return json.load(file)
```

```python
print("Started JSON exporting of git repos at {}".format(datetime.now()))
start_time = time()
export_git_repos_json('git_repos.json')
print("Finished JSON exporting of git repos at {} (took {} seconds)".format(datetime.now(), time() - start_time))
```

```python
print("Started importing JSON git repos at {}".format(datetime.now()))
start_time = time()
git_repos = import_git_repos_json('git_repos.json')
print("Finished importing JSON git repos at {} (took {} seconds)".format(datetime.now(), time() - start_time))
print(git_repos)
```
